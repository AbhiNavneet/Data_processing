```python
race_0 = data[data[:,2]==0, :]
race_1 = data[data[:,2]==1, :]
race_2 = data[data[:,2]==2, :]
race_3 = data[data[:,2]==3, :]
race_4 = data[data[:,2]==4, :]
```

```python
len_0 = len(race_0)
len_1 = len(race_1)
len_2 = len(race_2)
len_3 = len(race_3)
len_4 = len(race_4)
```

```python
print(len_0)
print(len_1)
print(len_2)
print(len_3)
print(len_4)
```

```python
smallest_pop = np.min(race_coll)
```
